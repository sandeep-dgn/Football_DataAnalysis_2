### Dataset Fix: Column Length Mismatch
The initial loading attempt failed due to a mismatch between the expected number of players and the actual columns in the CSV. 
I updated the loader to:
* Dynamically detect the number of columns.
* Apply team-specific prefixes (`Home_` or `Away_`).
* Explicitly map the last two columns to `Ball_X` and `Ball_Y`.


### Renaming Tracking Columns to X and Y


```python
def load_metrica_data(file_path, team_name):
    # 1. Load just the headers to find player numbers
    # Row 0 has "Home/Away", Row 1 has the Player Numbers (11, 1, 2, etc.)
    player_row = pd.read_csv(file_path, skiprows=1, nrows=0)
    raw_columns = player_row.columns
    
    # 2. Load the actual data (starting from row 3)
    df = pd.read_csv(file_path, skiprows=3, header=None)
    
    # 3. Build the column list dynamically
    colnames = ['Period', 'Frame', 'Time [s]']
    
    # We loop through the columns of the raw file to build our name list
    # We skip the first 3 (Period, Frame, Time)
    for i in range(3, len(raw_columns)):
        col_name = raw_columns[i]
        if "Unnamed" not in col_name:
            # This is a new player start
            last_player = col_name
            colnames.append(f"{team_name}_{last_player}_X")
        else:
            # This is the Y coordinate for the last player we found
            colnames.append(f"{team_name}_{last_player}_Y")
    
    # 4. The last two columns in Metrica data are ALWAYS the Ball
    colnames[-2] = "Ball_X"
    colnames[-1] = "Ball_Y"
    
    # Safety Check: If there is still a mismatch, fill the rest with 'Unknown'
    if len(colnames) < len(df.columns):
        for i in range(len(colnames), len(df.columns)):
            colnames.append(f"Extra_{i}")
            
    df.columns = colnames
    return df
```

```python
df_home=load_metrica_data(path_tracking_home, "Home")
df_away=load_metrica_data(path_tracking_away, "Away")
```

### Field Dimension Configuration
Pitch sizes vary by stadium. By defining variables, we ensure the 
analysis remains flexible for different datasets. 
For Metrica Sample Game 1, the standard dimensions are 105m x 68m.

while the given dataset has normalize the data to [0,1] we need to convert that to meter 

```python
# Constants for this specific match
PITCH_LENGTH = 105.0
PITCH_WIDTH = 68.0
```

```python


# You can now update the conversion function to use these variables
def convert_to_meters_flexible(df, length, width):
    x_columns = [c for c in df.columns if '_X' in c or 'Start X' in c or 'End X' in c]
    y_columns = [c for c in df.columns if '_Y' in c or 'Start Y' in c or 'End Y' in c]
    
    # Apply transformation
    df[x_columns] = (df[x_columns] - 0.5) * length
    df[y_columns] = (df[y_columns] - 0.5) * width
    
    return df
```

```python
df_home=convert_to_meters_flexible(df_home, PITCH_LENGTH, PITCH_WIDTH)
df_away=convert_to_meters_flexible(df_away, PITCH_LENGTH, PITCH_WIDTH)
```

### Quantifying Player Presence (Missing Data)
Since the dataset includes the entire match squad, many columns contain `NaN` values 
for periods when players were on the bench. I am calculating the 'Presence' 
percentage for each player to identify starters vs. substitutes.

```python
# 1. Get a list of all columns that end in '_X' for both teams
home_x_cols = [col for col in df_home.columns if col.endswith('_X') and 'Ball' not in col]
away_x_cols = [col for col in df_away.columns if col.endswith('_X') and 'Ball' not in col]

# 2. Check nulls ONLY in those specific columns
home_presence_specific = (1 - df_home[home_x_cols].isnull().mean()) * 100
away_presence_specific = (1 - df_away[away_x_cols].isnull().mean()) * 100

```

```python
# 3. Print the results
print("--- HOME PLAYER PRESENCE (X-COORD ONLY) ---")
print(home_presence_specific[home_presence_specific > 0].sort_values(ascending=False))

```

```python
print("\n--- AWAY PLAYER PRESENCE (X-COORD ONLY) ---")
print(away_presence_specific[away_presence_specific > 0].sort_values(ascending=False))
```

### Categorizing passes 

looking at the player coordinates and ball coordinates we have defined/categorized the pases into long pass,pass in oppsiton half,etc

```python
def add_event_zones(df):
    # 'Own Half' vs 'Opponent Half' based on X = 0
    # Assuming Home team attacks left to right (X > 0 is Opponent Half)
    df['In_Opponent_Half'] = df['Start X'] > 0
    
    # 'Long Ball' check (if distance between Start X and End X > 30 meters)
    df['Pass_Distance'] = ((df['End X'] - df['Start X'])**2 + (df['End Y'] - df['Start Y'])**2)**0.5
    df['Is_Long_Ball'] = df['Pass_Distance'] > 30
    
    return df

```
